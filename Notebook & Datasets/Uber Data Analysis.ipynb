## Data Preprocessing

### I. Rides 

```python
rides.info()
```

```python
rides.head(1)
```

```python
rides['ride_date'] = pd.to_datetime(rides['ride_date'], errors='coerce')

rides['start_datetime'] = pd.to_datetime(
    rides['ride_date'].astype(str) + ' ' + rides['start_time'],
    errors='coerce'
)

rides['end_datetime'] = pd.to_datetime(
    rides['ride_date'].astype(str) + ' ' + rides['end_time'],
    errors='coerce'
)
```

```python
# identify midnight-crossing rides
mask = rides['end_datetime'] < rides['start_datetime']

# add 1 day to end_datetime for those rides
rides.loc[mask, 'end_datetime'] = rides.loc[mask, 'end_datetime'] + pd.Timedelta(days=1)
```

```python
rides['ride_duration_min'] = (
    (rides['end_datetime'] - rides['start_datetime'])
    .dt.total_seconds() / 60
)
```

```python
cat_cols = ['payment_method', 'ride_status']

for col in cat_cols:
    rides[col] = rides[col].astype(str).str.strip().str.lower()
```
